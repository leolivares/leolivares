```python
import requests
from bs4 import BeautifulSoup

URL = 'http://www.basenotes.net/fragrancedirectory/'

data = requests.get(URL)
```

```python
content = data.text
```

```python
soup = BeautifulSoup(content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
directories_url = soup.select('a[href^="http://www.basenotes.net/fragrancedirectory/?i="]')
directories_url = [x["href"] for x in directories_url]
directories_url
```

```python
def get_perfumes_url(d_url):
    directory_data = requests.get(d_url)
    
    content = directory_data.text
    soup = BeautifulSoup(content, 'html.parser')
    
    urls = soup.select('a[href^="http://www.basenotes.net/ID"]')[:10]
    urls = [x["href"] for x in urls]
    return urls
```

```python
perfume_urls = list()
for d_url in directories_url:
    perfume_urls.append(get_perfumes_url(d_url))
perfume_urls
```

```python
def get_attributes(perfume_url):
    perfume_attr = dict()
    perfume_data = requests.get(perfume_url)
    
    content = perfume_data.text
    soup = BeautifulSoup(content, 'html.parser')
    
    name = soup.select('span[itemprop="name"]')[0].text
    try:
        year = soup.select('span[itemprop="releaseDate"]')[0].text
    except IndexError:
        year = "N/A"
    gender = soup.select('td[colspan="2"]')[0].text
    availability = soup.select('td[style="width:33px;"]')[1].text
    try:
        rating = soup.select('meta[itemprop="ratingValue"]')[0]['content']
    except IndexError as err:
        rating = "No ratings"
    perfume_attr['name'] = name
    perfume_attr['year'] = year
    perfume_attr['gender'] = gender
    perfume_attr['availability'] = availability
    perfume_attr['rating'] =  rating
    return perfume_attr
```

```python
perfume_list = list()
for group in perfume_urls:
    for url in group:
        perfume_list.append(get_attributes(url))

print(len(perfume_list))
print(perfume_list[:10])
```
