```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
```

```python
df = pd.read_csv('FATS_GAIA.dat')
```

```python
msk = np.random.rand(len(df)) < 0.8
train_df = df[msk]
test_df = df[~msk]
```

```python
X = train_df.drop('PeriodLS', 1)
X = X.drop('Class', 1)
y = train_df['PeriodLS']
```

```python
regr = DecisionTreeRegressor(max_depth=10)
regr.fit(X, y)
```

```python
X_test = test_df.drop('PeriodLS', 1)
X_test = X_test.drop('Class', 1)
y_test = test_df['PeriodLS']
```

```python
y_1 = regr.predict(X_test)
```

```python
y_1[4]
```

```python
info = {'expected': y_test, 'predicted': y_1}
results = pd.DataFrame(data=info)
```

```python
results
```

```python
regr.score(X_test, y_test, sample_weight=None)
```
